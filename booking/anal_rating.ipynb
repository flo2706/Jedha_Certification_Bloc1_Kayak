```python
import pandas as pd
import json
```

```python
with open("hotels_descriptions.json", "r", encoding="utf-8") as f:
    data = json.load(f)

df_hotels = pd.DataFrame(data)

print(df_hotels.head())
```

```python
print("Nbr of NUll:")
print(df_hotels.isnull().sum())
```

```python
print("Rows containing at least one null value:")
print(df_hotels[df_hotels.isnull().any(axis=1)])
```

```python
print("Hotels without rating :")
print(df_hotels[df_hotels["rating"].isna()])
```
