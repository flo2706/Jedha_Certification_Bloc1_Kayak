```python
import pandas as pd
```

```python
df1 = pd.read_csv("coordinates_clean.csv")
df2 = pd.read_csv("hotels_bis.csv")  
df3 = pd.read_csv("weather_clean.csv")  
```

```python
df1_clean = df1[["city_id", "city_name"]].drop_duplicates()
df2_clean = df2[["city_id", "city_name"]].drop_duplicates()
df3_clean = df3[["city_id", "city_name"]].drop_duplicates()
```

```python
# Fusion on city_id
merged = df1_clean.merge(df2_clean, on="city_id", suffixes=("_1", "_2")) \
                  .merge(df3_clean, on="city_id") \
                  .rename(columns={"city_name": "city_name_3"})

# Check 
merged["match_1_2"] = merged["city_name_1"] == merged["city_name_2"]
merged["match_1_3"] = merged["city_name_1"] == merged["city_name_3"]

# print dismatch
errors = merged[(~merged["match_1_2"]) | (~merged["match_1_3"])]
print("Dismatch :")
print(errors[["city_id", "city_name_1", "city_name_2", "city_name_3"]])
```

```python
df_hotels = pd.read_csv("hotels_bis.csv")
df_ref = pd.read_csv("coordinates_clean.csv")[["city_id", "city_name"]]

df_hotels = df_hotels.drop(columns=["city_name"], errors="ignore")

# Merge 
df_hotels = df_hotels.merge(df_ref, on="city_id", how="left")

# Check
print(df_hotels[["city_id", "city_name"]].drop_duplicates())
```

```python
df1 = pd.read_csv("coordinates_clean.csv")  
df2 = pd.read_csv("hotels_bis.csv")  
df3 = pd.read_csv("weather_clean.csv")  
```

```python
df1_clean = df1[["city_id", "city_name"]].drop_duplicates()
df2_clean = df2[["city_id", "city_name"]].drop_duplicates()
df3_clean = df3[["city_id", "city_name"]].drop_duplicates()
```

```python

merged = df1_clean.merge(df2_clean, on="city_id", suffixes=("_1", "_2")) \
                  .merge(df3_clean, on="city_id") \
                  .rename(columns={"city_name": "city_name_3"})

merged["match_1_2"] = merged["city_name_1"] == merged["city_name_2"]
merged["match_1_3"] = merged["city_name_1"] == merged["city_name_3"]

errors = merged[(~merged["match_1_2"]) | (~merged["match_1_3"])]
print("🔍 Incohérences trouvées :")
print(errors[["city_id", "city_name_1", "city_name_2", "city_name_3"]])
```

```python

df_coords = pd.read_csv("coordinates_clean.csv")
df_weather = pd.read_csv("weather_clean.csv")
df_hotels = pd.read_csv("hotels_bis.csv")

df_cities = pd.merge(df_coords, df_weather, on="city_id", how="inner")

df_final = pd.merge(df_hotels, df_cities, on="city_id", how="left")

df_final.to_csv("hotels_weather_combined.csv", index=False)
```

```python
df = pd.read_csv("hotels_weather_combined.csv")

display(df.head())
```

```python
print(df.dtypes)
```

```python
columns_to_keep = [
    'hotel_id',
    'hotel_name',
    'hotel_url',
    'hotel_rating',
    'hotel_description',
    'latitude_x',       # latitude hôtel
    'longitude_x',      # longitude hôtel
    'city_id',
    'city_name',
    'latitude_y',       # latitude ville
    'longitude_y',      # longitude ville
    'temp_max',
    'humidity',
    'wind_speed',
    'clouds',
    'pop',
    'good_weather_score'
]
```

```python
df_clean = df[columns_to_keep]

df_clean = df_clean.rename(columns={
    'latitude_x': 'hotel_latitude',
    'longitude_x': 'hotel_longitude',
    'latitude_y': 'city_latitude',
    'longitude_y': 'city_longitude'
})
```

```python
top_destinations = (
    df_clean[['city_id', 'city_name', 'good_weather_score']]
    .drop_duplicates()
    .sort_values(by='good_weather_score', ascending=False)
    .head(5)
)
display(top_destinations)
```

```python
top_hotels = (
    df_clean[['hotel_name', 'hotel_rating', 'city_name', 'hotel_latitude', 'hotel_longitude']]
    .sort_values(by='hotel_rating', ascending=False)
    .head(20)
)
display(top_hotels)
```

```python
df_clean.to_csv("hotels_weather_final.csv", index=False, encoding='utf-8')
```
